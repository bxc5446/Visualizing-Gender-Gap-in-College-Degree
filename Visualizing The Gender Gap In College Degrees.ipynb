# Visualizing The Gender Gap In College Degrees

## Introduction

In this project we aim to visualize the gender gap in college majors over the years. The dataset is compiled by Randal Olson and is made available on his personal website [here](http://www.randalolson.com/). 

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degree = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
# Define the color blindness freindly colors
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255,128/255,14/255)
cb_opaque = (171/255,171/255,171/255)
```

## Compare Across All Degree Categories

We seperated the dataset into three categories:
- STEM
- Liberal Arts
- Others

The classification allows us to combine the data both simultaneous and seperately.

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

# Create a figure with a width of 16 and a height of 25
fig = plt.figure(figsize=(16,25))

## Generate the first column of line chart for STEM
for sp in range(0,18,3):
    cat_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degree['Year'], women_degree[stem_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degree['Year'], 100-women_degree[stem_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    ax.set_title(stem_cats[cat_index])
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    # Add horizontal line
    ax.axhline(50, c=cb_opaque,alpha=0.3)
    # Remove ticks
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom=False)
    # Remove spines
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    # Label the lines
    if cat_index == 0:
        ax.text(2003,85,'Women')
        ax.text(2003,15,'Men')
    elif cat_index == 5:
        ax.text(2003,10,'Women')
        ax.text(2003,85,'Men')
        ax.tick_params(labelbottom=True)

## Generate the second column of line chart for Liberal Art
for sp in range(1,14,3):
    cat_index = int((sp-1)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degree['Year'], women_degree[lib_arts_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degree['Year'], 100-women_degree[lib_arts_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    ax.set_title(lib_arts_cats[cat_index])
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    # Add horizontal line
    ax.axhline(50, c=cb_opaque,alpha=0.3)
    # Remove ticks
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom=False)
    # Remove spines
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    # Label the lines
    if cat_index == 0:
        ax.text(2003,85,'Women')
        ax.text(2003,15,'Men')
    elif cat_index == 4:
        ax.text(2003,40,'Women')
        ax.text(2003,60,'Men')
        ax.tick_params(labelbottom=True)

## Generate the third column of line chart for Others
for sp in range(2,20,3):
    cat_index = int((sp-2)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degree['Year'], women_degree[other_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degree['Year'], 100-women_degree[other_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    ax.set_title(other_cats[cat_index])
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    # Add horizontal line
    ax.axhline(50, c=cb_opaque,alpha=0.3)
    # Remove ticks
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom=False)
    # Remove spines
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    # Label the lines
    if cat_index == 0:
        ax.text(2003,90,'Women')
        ax.text(2003,5,'Men')
    elif cat_index == 5:
        ax.text(2003,30,'Women')
        ax.text(2003,70,'Men')
        ax.tick_params(labelbottom=True)

## Save image as png
plt.savefig('college_degree_gender_gap.png')
    
```
